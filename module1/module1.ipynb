```python
import pandas as pd

# Load the parquet file
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet')
df_zone = pd.read_csv('https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv')
```

```python
start_date = pd.to_datetime('2025-11-01')
end_date = pd.to_datetime('2025-12-01')
filtered_df = df[(df['lpep_pickup_datetime'] >= start_date) & (df['lpep_pickup_datetime'] < end_date)]
len(filtered_df[(filtered_df['trip_distance'] <=1)])
```

```python
max_distance = df[(df['trip_distance'] <=100)]['trip_distance'].max()
df[(df['trip_distance'] == max_distance)]['lpep_pickup_datetime']
```

```python
dt = pd.to_datetime('2025-11-18')
df_18 = df[(df['lpep_pickup_datetime'].dt.normalize() == dt)]
zone = df_18.groupby('PULocationID')['total_amount'].sum().idxmax()
df_zone[(df_zone['LocationID'] == zone)]['Zone'].values[0]
```

```python
zone = 'East Harlem North'
LocationID = df_zone[(df_zone['Zone'] == zone)]['LocationID']
filtered_df = df[(df['lpep_pickup_datetime'] >= start_date) & (df['lpep_pickup_datetime'] < end_date) & (df['PULocationID'] == LocationID.values[0])]
filtered_df
max_tip = filtered_df['tip_amount'].max()
do_zone = filtered_df[(filtered_df['tip_amount'] == max_tip)]['DOLocationID']
df_zone[(df_zone['LocationID'] == do_zone.values[0])]['Zone'].values[0]
```
